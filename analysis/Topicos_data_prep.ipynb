> ### #01 - NA's:

```
# NA's por colunas:
qtd_na_por_col_market = pd.DataFrame(market.isna().sum().sort_values(ascending=False).reset_index())
qtd_na_por_col_market.columns = ['Coluna', 'Qtd_NA']
qtd_na_por_col_market['Qtd_nao_NA'] = (len(market) - qtd_na_por_col_market['Qtd_NA'])
qtd_na_por_col_market['Percentual_de_NA'] = ((qtd_na_por_col_market['Qtd_NA'] * 100) / len(market)).round(2)
qtd_na_por_col_market.T
```

Há 5 colunas 100% NA:

```
print(qtd_na_por_col_market['Coluna'].loc[qtd_na_por_col_market['Percentual_de_NA'] == 100])
print('\n','-'*35)
```

Decisões de tratamento a se pensar: 
  - Excluir as 100%;
  - Definir uma porcentagem máxima de NA para que a feature possa ser incluída no modelo. 
  
  Exemplo: Usar features com 40% ou menos de "Percentual_de_Na", ou seja, 60% ou mais dos registros devem estar preenchidos;
  - Inferir valores e preencher NA's.
