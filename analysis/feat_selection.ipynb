# Feature Importance with ExtraTreesClassifier
* 1. Fit an Extra Trees model to the data
* 2. Display the relative importance of each attribute
* 3. Select features with value greater than threshold 

###  Portfolio1

```python
# Fit an Extra Trees model to the data
model = ExtraTreesClassifier()
model.fit(portfolio1_data.drop('portfolio',axis=1)._get_numeric_data(), portfolio1_data['portfolio'])
# Display the relative importance of each attribute
importance = model.feature_importances_
print(importance)
```

```python
importance_df = pd.DataFrame(importance.reshape(1,-1), columns = portfolio1_data.drop('portfolio',axis=1)._get_numeric_data().columns)
importance_df[importance_df > 0.04].dropna(axis = 'columns').columns
#importance_df.head()
```
